## Gr√°ficos de correntes I1 e I2 em fun√ß√£o da frequ√™ncia

```python
from math import *
from cmath import *
from numpy import linalg
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. C√≥digo do modelo num√©rico

```python
R1 = 439.4e-3
R2 = 424.1e-3
L1 = 87.66e-6
L2 = 86.29e-6
C1 = C2 = 1e-6

RC = 300
f = 1/(2*np.pi*np.sqrt(C1*L1))
k = 0.5

w = 2*pi*f
M = k*sqrt(L1*L2)

XL1 = w * L1 * 1j
XL2 = w * L2 * 1j

XC1 = 1/(w * C1 * 1j)
XC2 = 1/(w * C2 * 1j)

XM = w * M * 1j

ZEQ = (RC * XC2)/(RC + XC2)

print("Frequ√™ncia", f)
print("W = ", w)
print("M = ", M)
print("XL1 = ", XL1)
print("XL2 = ", XL2)
print("XC1 = ", XC1)
print("XC2 = ", XC2)
print("XM = ", XM)
print("ZEQ = ", ZEQ)
```

```python
vp = 9
V1 = (2/np.pi)*vp
```

```python
def CalcularTransformador(v1, ZEQ):
    Z=np.array([[R1+XL1+XC1, -XM],[-XM, XL2+R2+ZEQ]])
    V=np.array([v1,0])
    I=np.dot(linalg.inv(Z),V)
    return I[0], I[1]
I1, I2 = CalcularTransformador(V1, ZEQ)
V2 = I2 * ZEQ
print(I1, I2)
print(V2)
```

#### 3. Constru√ß√£o de gr√°ficos de I1 e I2 em fun√ß√£o da frequ√™ncia

```python
fs = np.linspace(1, 100000, 100)
I1_list = []
I2_list = []

for f in fs:
    w = 2*np.pi*f
    XL1 = w * L1 * 1j
    XL2 = w * L2 * 1j
    XC1 = 1/(w * C1 * 1j)
    XC2 = 1/(w * C2 * 1j)
    XM = w * M * 1j

    I1, I2 = CalcularTransformador(V1, ZEQ)
    I1_list.append(np.abs(I1))
    I2_list.append(np.abs(I2))


plt.plot(fs, I1_list)
plt.plot(fs, I2_list)
plt.xlabel('Frequ√™ncia (Hz)')
plt.ylabel('Corrente (A)')
plt.legend(['I1', 'I2'])
plt.grid()
plt.show()
```

#### 4. Constru√ß√£o de um gr√°fico do m√≥dulo da tens√£o ùëâ2 pela frequ√™ncia de resson√¢ncia, para 6 diferentes indut√¢ncias

```python
L_values = [100e-6, 200e-6, 300e-6, 400e-6, 500e-6, 600e-6]
fr_values = np.linspace(0, 100e3, 1000) 
R1 = 439.4e-3
R2 = 424.1e-3
k = 0.5

for L in L_values:
    L1 = L2 = L
    V2_values = []
    for fr in fr_values:
        w = 2 * np.pi * fr
        if w == 0:
            C1 = C2 = np.inf
            XC1 = XC2 = 0
        else:
            C1 = C2 = 1 / ((w ** 2) * L1)
            XL1 = w * L1 * 1j
            XL2 = w * L2 * 1j
            XC1 = 1 / (w * C1 * 1j)
            XC2 = 1 / (w * C2 * 1j)
        M = k * np.sqrt(L1 * L2)
        XM = w * M * 1j
        ZEQ = (RC * XC2) / (RC + XC2)
        
        I1, I2 = CalcularTransformador(V1, ZEQ)
        V2 = I2 * ZEQ
        V2_values.append(abs(V2))

    plt.plot(fr_values, V2_values, label=f'L={L*1e6:.0f} ¬µH')

plt.xlabel("Frequ√™ncia de Resson√¢ncia (Hz)")
plt.ylabel("|V2| (V)")
plt.title("Gr√°fico de |V2| vs frequ√™ncia de resson√¢ncia para diferentes valores de L")
plt.legend()
plt.grid(True)
plt.show()
```

#### 5. Constru√ß√£o de gr√°ficos das pot√™ncias e da efici√™ncia em fun√ß√£o da frequ√™ncia

```python
R1 = 439.4e-3
R2 = 424.1e-3
C1 = C2 = 1e-6
L1 = 87.66e-6
L2 = 86.29e-6
fr_values = np.linspace(0, 100e3, 1000)
k = 0.5

P_in_list = []
P_out_list = []
efficiency_list = []

for fr in fr_values:
    w = 2 * np.pi * fr
    if w == 0:
        XC1 = XC2 = 0
    else:
        XL1 = w * L1 * 1j
        XL2 = w * L2 * 1j
        XC1 = 1 / (w * C1 * 1j)
        XC2 = 1 / (w * C2 * 1j)
    M = k * np.sqrt(L1 * L2)
    XM = w * M * 1j
    ZEQ = (RC * XC2) / (RC + XC2)

    I1, I2 = CalcularTransformador(V1, ZEQ)
    V2 = I2 * ZEQ
    P_in = V1 * I1.conjugate() / 2
    P_out = V2 * I2.conjugate() / 2

    if P_in != 0:
        efficiency = (P_out.real / P_in.real)*100
    else:
        efficiency = 0

    P_in_list.append(P_in.real)
    P_out_list.append(P_out.real)
    efficiency_list.append(efficiency)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.plot(fr_values, P_in_list, label='Pot√™ncia de Entrada')
ax1.plot(fr_values, P_out_list, label='Pot√™ncia de Sa√≠da')
ax1.set_xlabel('Frequ√™ncia (Hz)')
ax1.set_ylabel('Pot√™ncia (W)')
ax1.set_title('Pot√™ncia de Entrada e Sa√≠da vs Frequ√™ncia de Resson√¢ncia')
ax1.legend()
ax1.grid(True)

ax2.plot(fr_values, efficiency_list, label='Efici√™ncia', color='green')
ax2.set_xlabel('Frequ√™ncia (Hz)')
ax2.set_ylabel('Efici√™ncia (%)')
ax2.set_title('Efici√™ncia vs Frequ√™ncia de Resson√¢ncia')
ax2.grid(True)

#plt.tight_layout()
plt.show()
```

#### 6. Constru√ß√£o de um gr√°fico que relacione o valor de pico da tens√£o no secund√°rio (no resistor ùëÖùê∂) como coeficiente de acoplamento k

```python
k_values = np.linspace(0, 1, 500)
R1 = 439.4e-3
R2 = 424.1e-3
L1 = 87.66e-6
L2 = 86.29e-6
C1 = C2 = 1e-6
f = 1/(2*np.pi*np.sqrt(C1*L1))
w = 2*pi*f

V2_list = []

for k in k_values:
    XL1 = w * L1 * 1j
    XL2 = w * L2 * 1j
    XC1 = 1 / (w * C1 * 1j)
    XC2 = 1 / (w * C2 * 1j)
    M = k * sqrt(L1*L2)
    XM = w * M * 1j
    ZEQ = (RC * XC2)/(RC + XC2)

    I1, I2 = CalcularTransformador(V1, ZEQ)
    V2 = I2 * ZEQ
    V2_list.append(abs(V2))

k_max = k_values[np.argmax(V2_list)]
V2_max = np.max(V2_list)

plt.plot(k_values, V2_list, label='V2 (Pico)')
plt.xlabel('Coeficiente de Acoplamento (k)')
plt.ylabel('Tens√£o no Secund√°rio (V2 Pico)')
plt.title('V2 Pico vs Coeficiente de Acoplamento (k)')
plt.scatter(k_max, V2_max, color='red', zorder=5)
plt.annotate(f'M√°ximo V2 ({V2_max:.2f}) em k={k_max:.2f}', (k_max, V2_max), xytext=(0.2, V2_max-0.5), 
            arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.grid(True)
plt.legend()
plt.show()

# Para cada frequ√™ncia existe uam dist√¢ncia √≥tima para transmitir energia e vice-versa!!!
```

#### 7. Simula√ß√£o das pot√™ncias reais em fun√ß√£o de R1 (resist√™ncia no prim√°rio)

```python
R1 = 439.4e-3
R2 = 424.1e-3
L1 = 87.66e-6
L2 = 86.29e-6
C1 = C2 = 1e-6
k = 0.1
f = 1/(2*np.pi*np.sqrt(C1*L1))
R1_values = np.linspace(R1, R1 + 4.5, 100)

P_in_list = []
P_out_list = []
efficiency_list = []

for R in R1_values:
    R1 = R
    w = 2 * np.pi * f
    if w == 0:
        XC1 = XC2 = 0
    else:
        XL1 = w * L1 * 1j
        XL2 = w * L2 * 1j
        XC1 = 1 / (w * C1 * 1j)
        XC2 = 1 / (w * C2 * 1j)
    M = k * np.sqrt(L1 * L2)
    XM = w * M * 1j
    ZEQ = (RC * XC2) / (RC + XC2)

    I1, I2 = CalcularTransformador(V1, ZEQ)
    V2 = I2 * ZEQ
    P_in = V1 * I1.conjugate() / 2
    P_out = V2 * I2.conjugate() / 2

    if P_in != 0:
        efficiency = (P_out.real / P_in.real)*100
    else:
        efficiency = 0

    P_in_list.append(P_in.real)
    P_out_list.append(P_out.real)
    efficiency_list.append(efficiency)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
ax1.plot(R1_values, P_in_list, label='Pot√™ncia de Entrada')
ax1.plot(R1_values, P_out_list, label='Pot√™ncia de Sa√≠da')
ax1.set_xlabel('Resist√™ncia (Ohm)')
ax1.set_ylabel('Pot√™ncia (W)')
ax1.set_title('Pot√™ncia R1 Entrada e Sa√≠da vs R1')
ax1.legend()
ax1.grid(True)

ax2.plot(R1_values, efficiency_list, label='Efici√™ncia', color='green')
ax2.set_xlabel('Resist√™ncia (Ohms)')
ax2.set_ylabel('Efici√™ncia (%)')
ax2.set_title('Efici√™ncia vs R1')
ax2.grid(True)

#plt.tight_layout()
plt.show()
```
